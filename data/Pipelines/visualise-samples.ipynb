```python
import pydicom
import matplotlib.pyplot as plt
import numpy as np
```

```python
def visulise_pet(path):

    # read DICOM
    dcm = pydicom.dcmread(path)
    img = dcm.pixel_array.astype(np.float32)

    # normalize image to [0,1] for visualization
    img_norm = (img - np.min(img)) / (np.max(img) - np.min(img))

    # plot PET image
    plt.figure(figsize=(6,6))
    # 'hot' colormap highlights uptake
    plt.imshow(img_norm, cmap='hot')
    plt.title(f"PET: Patient ID {dcm.PatientID}")
    plt.axis('off')
    plt.colorbar(label='Normalized Intensity')
    plt.show()
```

```python
def visualise_ct(path):
    # read DICOM
    dcm = pydicom.dcmread(path)

    # extract pixel data
    img = dcm.pixel_array.astype(float)

    # normalize to [0, 1] for display
    img_norm = (img - np.min(img)) / (np.max(img) - np.min(img))

    # Plot
    plt.figure(figsize=(6,6))
    plt.imshow(img_norm, cmap='gray')
    plt.title(f"Patient ID: {dcm.PatientID}")
    plt.axis('off')
    plt.show()
```

#### Lung-PET-CT-Dx PET

```python
# path to PET DICOM
path = "../raw/PET/A/Lung-PET-CT-Dx-A0210/4-051.dcm"

visulise_pet(path)
```
