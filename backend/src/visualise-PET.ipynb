```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize_pet_slices(path):

    # load the PET volume into 3D array
    pet_volume = np.load(path)

    print(f"Shape: {pet_volume.shape}")
    print(f"Data type: {pet_volume.dtype}")
    print(f"Value range: {pet_volume.min():.4f} → {pet_volume.max():.4f}")

    # visualize the first 16 slices
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    axes = axes.flatten()

    for i in range(4):
        # get the i-th slice
        slice_data = pet_volume[i]  # shape: [H, W]
        
        # display the slice
        im = axes[i].imshow(slice_data, cmap='hot', interpolation='nearest')
        axes[i].set_title(f'Slice {i+1}\nShape: {slice_data.shape}')
        axes[i].axis('off')
        
        # add colorbar for each slice
        plt.colorbar(im, ax=axes[i], fraction=0.046, pad=0.04)

    plt.suptitle('First 4 Slices', fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
path = '../../data/preprocessed/PET/up_threshold_0.8/A/Lung-PET-CT-Dx-A0223.npy'

visualize_pet_slices(path)
```
