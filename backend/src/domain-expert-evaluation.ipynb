### Results

```python
# data
questions = [f"Q{i}" for i in range(1, 11)]
e1 = [8, 8, 8, 6, 7, 8, 8, 8, 8, 9]
e2 = [7, 7, 8, 7, 7, 8, 6, 7, 8, 8]
e3 = [8, 8, 6, 7, 6, 8, 7, 7, 6, 9]

# angles
N = len(questions)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

# close the loop
e1 += e1[:1]
e2 += e2[:1]
e3 += e3[:1]

c1 = "#8CB9FF"  # light blue
c2 = "#FFC79A"  # light peach
c3 = "#92D9A7"  # light green

# plot
plt.figure(figsize=(8, 8), facecolor="#FFFDF8")
ax = plt.subplot(111, polar=True)

# lines + filled regions
ax.plot(angles, e1, color=c1, linewidth=2.6, marker="o", markersize=5, label="Evaluator 1")
ax.fill(angles, e1, color=c1, alpha=0.20)

ax.plot(angles, e2, color=c2, linewidth=2.6, marker="o", markersize=5, label="Evaluator 2")
ax.fill(angles, e2, color=c2, alpha=0.20)

ax.plot(angles, e3, color=c3, linewidth=2.6, marker="o", markersize=5, label="Evaluator 3")
ax.fill(angles, e3, color=c3, alpha=0.20)

# axes styling
ax.set_xticks(angles[:-1])
ax.set_xticklabels(questions, fontsize=16, color="#1E293B")
ax.set_yticklabels([])
ax.grid(color="#B8C4D6", alpha=0.6)

ax.set_title("Domain Expert Evalutation", pad=20, color="#0F172A", fontsize=20)
ax.legend(loc="upper right", bbox_to_anchor=(1.25, 1.1), frameon=False, fontsize=16)

plt.tight_layout()
plt.show()
```
